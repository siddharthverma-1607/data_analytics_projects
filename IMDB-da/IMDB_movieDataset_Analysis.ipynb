```python
# 20. Classify Movies Based on Ratings [Excellent, Good and Average]
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_category']=data['Rating'].apply(rating)
```

```python
# 22. Find unique values from Genre
allGenre_temp = []
for value in data['Genre']:
    allGenre_temp.append(value.split(','))
    
allGenre_temp[:5]
```

```python
unique_genre = []
for item in allGenre_temp:
    for genre in item:
        unique_genre.append(genre)

allGenre_1d = unique_genre
unique_genre = list(set(unique_genre))
unique_genre
```
